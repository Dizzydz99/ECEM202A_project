```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
data_filename = "sensor_calibration.csv"
df = pd.read_csv(data_filename)
```

```python
sample_time = 300 # sampled for 5 mins
num_sample = df.shape[0]
AcceX = df['ax']
AcceY = df['ay']
AcceZ = df['az']

print('AcceX offset:' + str(np.mean(AcceX)))
print('AcceY offset:' + str(np.mean(AcceY)))
print('AcceZ offset:' + str(np.mean(AcceZ - 1)))

x_axis = [x / num_sample * sample_time for x in range(num_sample)]

plt.plot(x_axis, AcceX, label='AcceX')
plt.plot(x_axis, AcceY, label='AcceY')
plt.plot(x_axis, (AcceZ - 1), label='AcceZ')
plt.xlabel('Time(sec)')
plt.ylabel('Acceleration(G)')
plt.title('Stationary Accelerometer reading vs Time')
plt.legend()
plt.grid()
plt.show()
```

```python
MagX = df['mx']
MagY = df['my']
MagZ = df['mz']
plt.plot(x_axis, MagX, label='MagX')
plt.plot(x_axis, MagY, label='MagY')
plt.plot(x_axis, MagZ, label='MagZ')
plt.xlabel('Time(sec)')
plt.ylabel('Magnetic')
plt.title('Stationary Magnetic reading vs Time')
plt.legend()
plt.grid()
plt.show()
```

```python
GyroX = df['gx']
GyroY = df['gy']
GyroZ = df['gz']

print('GyroX offset:' + str(np.mean(GyroX)))
print('GyroY offset:' + str(np.mean(GyroY)))
print('GyroZ offset:' + str(np.mean(GyroZ)))

plt.plot(x_axis, GyroX, label='GyroX')
plt.plot(x_axis, GyroY, label='GyroY')
plt.plot(x_axis, GyroZ, label='GyroZ')

plt.xlabel('Time(sec)')
plt.ylabel('Gyro')
plt.title('Stationary Gyro reading vs Time')
plt.legend()
plt.grid()
plt.show()
```

```python
ptich = df['pitch']
roll = df['roll']
yaw = df['yaw']
plt.plot(x_axis, ptich, label='ptich')
plt.plot(x_axis, roll, label='roll')
plt.plot(x_axis, yaw, label='yaw')
plt.xlabel('Time(sec)')
plt.ylabel('Angle(rad)')
plt.title('Stationary inclination angle vs Time')
plt.legend()
plt.grid()
plt.show()
```
